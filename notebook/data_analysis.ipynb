```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/india_census_2011.csv")

df.head()
```

```python
print("Shape:", df.shape)

df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["Literacy_Rate"] = (
    df["Literate"] / df["Population"]
) * 100

df["Female_Ratio"] = (
    df["Female"] / df["Male"]
) * 1000

df["Worker_Rate"] = (
    df["Workers"] / df["Population"]
) * 100

df["Graduate_Rate"] = (
    df["Graduate_Education"] / df["Population"]
) * 100
```

```python
corr_cols = [
    "Literacy_Rate",
    "Worker_Rate",
    "Female_Ratio",
    "Graduate_Rate"
]

corr = df[corr_cols].corr()

plt.figure(figsize=(6,4))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Between Key Indicators")
plt.savefig("../figures/correlation.png")
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

if "Development_Score" not in df.columns:
    scaler = MinMaxScaler()

    metrics = ["Literacy_Rate", "Worker_Rate", "Graduate_Rate"]
    scaled = scaler.fit_transform(df[metrics])

    df["Development_Score"] = (
        scaled[:, 0] * 0.4
        + scaled[:, 1] * 0.3
        + scaled[:, 2] * 0.3
    )
```

```python
state_scores = (
    df.groupby("State_name")["Development_Score"]
      .mean()
      .sort_values()
)

plt.figure(figsize=(10,8))

sns.barplot(
    x=state_scores.values,
    y=state_scores.index
)

plt.title("Average Participation Score by State")
plt.savefig("../figures/state_scores.png")
plt.show()
```

WELFARE PARTICIPATION SCORE

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

metrics = [
  "Literacy_Rate",
  "Worker_Rate",
  "Graduate_Rate"
]

scaled = scaler.fit_transform(
  df[metrics]
)
df["Development_Score"] = (
    scaled[:,0] * 0.4
    +
    scaled[:,1] * 0.3
    +
    scaled[:,2] * 0.3
)
```

DISTRICT SEGMENTATION MODEL

```python
df["Segment"] = pd.qcut(
    df["Development_Score"],
    q=4,
    labels=[
        "Critical",
        "Low",
        "Medium",
        "High"
    ]
)
```

```python
df["Segment"].value_counts()
```

```python
sns.countplot(
    y=df["Segment"],
    order=[
        "High",
        "Medium",
        "Low",
        "Critical"
    ]
)

plt.title("District Segmentation")
plt.savefig("../figures/segmentation.png")
plt.show()
```

```python
bottom10 = (
    df.sort_values(
        "Development_Score"
    )
    .head(10)
)

bottom10[
    [
        "State_name",
        "District_name",
        "Development_Score"
    ]
]
```

```python
intervention = bottom10[
    [
        "State_name",
        "District_name",
        "Development_Score"
    ]
].copy()

intervention["Recommendation"] = (
    "Awareness Campaign + Enrollment Support"
)

intervention
```

```python
intervention.to_csv(
    "../report/intervention_map.csv",
    index=False
)
```

```python
corr_cols = [
    "Literacy_Rate",
    "Worker_Rate",
    "Graduate_Rate",
    "Development_Score"
]

plt.figure(figsize=(8,6))

sns.heatmap(
    df[corr_cols].corr(),
    annot=True,
    cmap="Blues"
)

plt.title("Correlation Analysis")
plt.savefig("../figures/correlation_heatmap.png")
plt.show()
```

```python
def recommend(row):

    if row["Literacy_Rate"] < 50:
        return "Literacy & Awareness Campaign"

    elif row["Graduate_Rate"] < 3:
        return "Educational Outreach Program"

    elif row["Worker_Rate"] < 30:
        return "Employment & Welfare Enrollment Drive"

    else:
        return "Targeted Welfare Awareness"

intervention["Recommendation"] = (
    bottom10.apply(
        recommend,
        axis=1
    )
)
```

## Limitations

This analysis uses demographic and educational indicators as proxies for welfare participation because district-level scheme uptake data was unavailable.

Future work can integrate PM-KISAN, Ayushman Bharat, PMUY, or MGNREGA enrollment datasets to directly measure welfare participation and improve intervention targeting.
